```python
import pandas as pd
```

```python
data = pd.read_csv('FullDF_duplicated.csv',encoding='utf-8-sig',index_col=0)
data
```

```python
data = data.drop('리뷰',axis=1)
```

```python
df_table = data.set_index(["작성자",'0']).unstack()
df_table
```

```python
df_table.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ibc = cosine_similarity(df_table)
ibc = pd.DataFrame(data=ibc, index=df_table.index, columns=df_table.index)
ibc
```

```python
def get_item_based_collabor(item):
    return ibc[item].sort_values(ascending=False)[:6]
```

```python
get_item_based_collabor('섬아저씨')
```
